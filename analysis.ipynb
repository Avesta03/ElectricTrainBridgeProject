```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/avestaafshari-mehr/Documents/ElectricTrainBridgeProject/data.csv')
```

```python
gauge_size = df['gaugesize']
average_velocity = df['averagevelocity']
velocity_error = df['velocityerror']
```

```python
plt.figure(figsize=(4, 3))
plt.errorbar(gauge_size, average_velocity, yerr=velocity_error, fmt='o', ecolor='black', capsize=5, capthick=2)
plt.xlabel('Diameter of wire (mm)')
plt.ylabel('Average Velocity (m/s)')
plt.grid(False)
plt.show()
```
